### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# Resize your Graph (dpi specifies pixels per inch, Use 300 while saving)
plt.figure(figsize=(5,3), dpi=200)


# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='blue', linewidth=2, linestyle='--', marker='.', markersize=10, markeredgecolor='pink')


# Shorthand Notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'y*--', label='2x')


# Select intervals to plot points at
x2 = np.arange(0,4.5,0.5)


# Plot part of the graph as Line
plt.plot(x2[:5], x2[:5]**2, 'r', label='x^2')


# Plot remainder of graph as Dot
plt.plot(x2[4:], x2[4:]**2, 'r--', label='x^2')


# Add a title (specify font parameters with fontdict)
plt.title('First Graph !!!', fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})


# X & Y labels
plt.xlabel('X-axis', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y-axis')


# X, Y axis Tickmarks (scale of your graph)
#plt.xticks([0,1,1.4,1.8,3])
#plt.yticks([2,4,6.5,7.5,8])


# Add a legend
plt.legend()


# Save figure
plt.savefig('mygraph.png', dpi=300)


# Show plot
plt.show()
```

### Bar Graph

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]


plt.figure(figsize=(4,4))


bars = plt.bar(labels, values)


bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')


plt.show()
```
